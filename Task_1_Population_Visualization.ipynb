```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv("population.csv", skiprows=4)

# Select year
year = "2023"

# Keep only Country Name and population for 2023
population = df[["Country Name", year]]

# Remove missing values
population = population.dropna()

# Get top 10 entries by population
top10 = population.sort_values(by=year, ascending=False).head(10)

# Create bar chart
plt.figure(figsize=(12, 6))
plt.bar(top10["Country Name"], top10[year])

plt.title("Top 10 Population Categories (2023)")
plt.xlabel("Country / Region")
plt.ylabel("Population")
plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```
